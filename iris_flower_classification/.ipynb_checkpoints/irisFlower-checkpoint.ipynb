# Splitting the Dataset into the Training set and Test set

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = 1)
print("Training set: {} and Test set: {}".format(X_train.shape, X_test.shape))
```

```python
X_train
```

```python
y_train
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logistic_regressor = LogisticRegression()
logistic_regressor.fit(X_train, y_train)
```

```python
y_pred_lgr = logistic_regressor.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("---- Logistic Regression - Model Score ----")
print("Accuracy Score:",accuracy_score(y_test,y_pred_lgr))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
decision_classifier = DecisionTreeClassifier(criterion='entropy',random_state=7)
decision_classifier.fit(X_train,y_train)
y_pred_dt =decision_classifier.predict(X_test)
print("---- Decision Tree Classifier - Model Score ----")
print("Accuracy Score:",accuracy_score(y_test,y_pred_dt))
```

# Support Vector Machine

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train,y_train)
y_pred_svm = svm.predict(X_test)
print("---- Support vector Machine - Model Score ----")
print("Accuracy Score:",accuracy_score(y_test,y_pred_svm))
```

# Stochastic Gradient Descent

```python
from sklearn.linear_model import SGDClassifier
sgd=SGDClassifier(loss="modified_huber",shuffle=True,random_state=100)
sgd.fit(X_train,y_train)
y_pred_sgd = sgd.predict(X_test)
print("---- Stochastic Gradient Descent - Model Score ----")
print("Accuracy Score: ",accuracy_score(y_test,y_pred_sgd))
```

# K - Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred_knn = knn.predict(X_test)
print("---- K-Nearest Neighbors - Model Score ----")
print("Accuracy Score: ",accuracy_score(y_test,y_pred_knn))
```

# GaussianNB

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,y_train)
y_pred_gnb = gnb.predict(X_test)
print("---- GaussianNB - Model Score ----")
print("Accuracy Score: ",accuracy_score(y_test,y_pred_gnb))
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
random_regressor = RandomForestClassifier()
random_regressor.fit(X_train, y_train)
y_pred_rfg = random_regressor.predict(X_test)
print("---- RandomForestClassifer - Model Score ----")
print("Accuracy Score: ",accuracy_score(y_test,y_pred_rfg))
```
