# Task 4 - Image to Pencil Sketch

# Importing Libraries

```python
import cv2
import numpy as np
from PIL import Image
from IPython.display import display
```

# Loading the Image

```python
image = cv2.imread('flower.jpg', 1)
image = cv2.resize(image, (640, 400))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

# Converting image into GrayScale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

# Inverting the Image

```python
inverted_image = np.invert(gray_image)
display(Image.fromarray(inverted_image))
```

# Average Blur the Image

```python
AvgBlur = cv2.blur(inverted_image, (5, 5))
display(Image.fromarray(AvgBlur))
```

# Smoothing the Image

```python
img_smooth = cv2.GaussianBlur(inverted_image, (5, 5), 0, 0)
display(Image.fromarray(img_smooth))
```

```python
inverted_image = 255 - gray_image
display(Image.fromarray(inverted_image))
```

# Gaussian Blur

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
display(Image.fromarray(blurred))
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
display(Image.fromarray(pencil_sketch))
```

# Comparing the Original Image with Pencil Sketch

```python
display(Image.fromarray(image))
display(Image.fromarray(pencil_sketch))
```
